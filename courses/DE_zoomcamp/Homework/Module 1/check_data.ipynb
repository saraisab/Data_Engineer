## Connection string

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

Creates the tables in the db

```python
df_green.head(n=0).to_sql(name='green_taxi_data', con=engine, if_exists='replace')
df_zone.head(n=0).to_sql(name='zone', con=engine, if_exists='replace')
```

Now, we check that the tables are created in the db

Imports the csv to the new table green_taxi_data

```python
%time df_green.to_sql(name='green_taxi_data', con=engine, if_exists='append')
```

Imports the zone csv to the new table zone

```python
%time df_zone.to_sql(name='zone', con=engine, if_exists='append')
```
